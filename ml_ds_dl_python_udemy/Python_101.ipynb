```python
# Correlation & Covariance
# Negative value shows inverse relation between features (correlation)
# Positive value shows direct relation between features (correlation)
# Correlation range is -1 to +1
```

```python
def de_mean(x):
    xmean = np.mean(x) # mean() is numpy function
    return [a - xmean for a in x]
```

```python
def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1) # dot() is numpy function
```

```python
page_speeds = np.random.normal(3.0, 1.0, 1000)
purchase_amount = np.random.normal(50.0, 10.0, 1000)
```

```python
covariance(page_speeds, purchase_amount)
```

```python
def correlation(x, y):
    std_dev_x = np.std(x)
    std_dev_y = np.std(y)
    
    return covariance(x, y) / std_dev_x / std_dev_y
```

```python
correlation(page_speeds, purchase_amount)
```

```python
# Conditional Probability
```

```python
from numpy import random
random.seed(0)
```

```python
totals = {20:0, 30:0, 40:0, 50:0, 60:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0}

total_purchases = 0
sample_size = 100000
```

```python
for _ in range(sample_size):
    age_decade = random.choice([20, 30, 40, 50, 60])
#     purchase_probability = float(age_decade) / 100.0
    purchase_probability = 0.10
    totals[age_decade] += 1
    if random.random() < purchase_probability:
        total_purchases += 1
        purchases[age_decade] += 1
    
```

```python
totals
```

```python
purchases
```

```python
total_purchases
```

```python
# Probability to purchase given that a person is in 30s ...
PEF = float(purchases[30]) / float(totals[30])
PEF
```

```python
# Probability of a person being in 30s ...
PF = float(totals[30]) / sample_size
PF
```

```python
# Probability of buying something
PE = total_purchases / sample_size
PE
```

```python
# P(E, F)
P = float(purchases[30]) / sample_size
P
```

```python
PE * PF
```

```python
# Hence proved that E and F are dependent on each other
```

```python
# Calculating conditional probability using equation -
P / PF
```
